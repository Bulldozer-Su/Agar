```python
import math
import numpy as np
import pandas as pd
import copy

# initialize parameters
total_time = 0
x_initial = 0
y_initial = 0
filename = 'data.txt'  
id = []
x = []
y = []
size = []
order = []

avgMax = 0
avgMaxPosition = 0
locateX = 0
locateY = 0
sumCount = 10
result_temp = 0
resultPoints = []

result_shortest = 0
result_shortest_temp = 0
value_temp = 0
data_temp = []
locateX_temp = 0
locateY_temp = 0
avgMin = 999999
avgMinPosition = 0
result_shortest_points = []

#extract data from the data.txt file
with open(filename, 'r') as f:
    for lines in f.readlines():
        lines = lines.strip('\n') 
        id_tmp, x_tmp, y_tmp, size_tmp = [float(i) for i in
                                          lines.split(',')]  
        id.append(id_tmp)  
        x.append(x_tmp)
        y.append(y_tmp)
        size.append(size_tmp)
    pass
id = np.array(id)  
x = np.array(x)
y = np.array(y)
size = np.array(size)
# reshape data
id = id.reshape(len(id), -1)
x = x.reshape(len(x), -1)
y = y.reshape(len(y), -1)
size = size.reshape(len(size), -1)

# transform into matrix
data = np.hstack((id, x, y, size))

for i in range(len(data) - 1):
    #get the most valuable cell
    for j in range(len(data) - 1):
        result = math.sqrt(math.pow(data[j][1] - locateX, 2) + math.pow(data[j][2] - locateY, 2))
        result_avg=data[j][3] / result
        if avgMax < result_avg and sumCount > data[j][3]: # catch the most valuable cell
            avgMax = result_avg
            result_temp = result # save the distance
            avgMaxPosition = j # save the index
    data_temp = copy.deepcopy(data)
    # figure out whether there is a better sequence
    while result_shortest <= result_temp:
        # figure out the nearest cell
        for k in range(len(data_temp) - 1):
            result = math.sqrt(math.pow(data_temp[k][1] - locateX_temp, 2) + math.pow(data_temp[k][2] - locateY_temp, 2))
            if avgMin > result and sumCount > data_temp[k][3]:
                avgMin = result # save the distance
                avgMinPosition = k # save the index
        result_shortest += avgMin # accumulate distance 
        #determine if ‘result_shortest’ is bigger than ‘result’
        if result_shortest <= result_temp:
            value_temp += data_temp[avgMinPosition][3]
            locateX_temp = data_temp[avgMinPosition][1]
            locateY_temp = data_temp[avgMinPosition][2]
            result_shortest_points.append((data_temp[avgMinPosition][0],data_temp[avgMinPosition][1],data_temp[avgMinPosition][2],data_temp[avgMinPosition][3]))
            data_temp = np.delete(data_temp, avgMinPosition, 0)
            result_shortest_temp = avgMin
            avgMin = 999999
    result_shortest -= avgMin
    # if its size is bigger than the most valuable cell, add this sequence into our eaten cells list
    if value_temp >= data[avgMaxPosition][3]:
        total_time += result_shortest
        if total_time > 10000:
            total_time -= result_shortest
            break
        sumCount += value_temp
        data = copy.deepcopy(data_temp)
        locateX = locateX_temp
        locateY = locateY_temp
        resultPoints.extend(result_shortest_points)
    else: # If not, drop this sequence then add the most valuable cell into the eaten cells list 
        total_time += result_temp
        if total_time > 10000:
            total_time -= result_temp
            break
        sumCount += data[avgMaxPosition][3]
        locateX = data[avgMaxPosition][1]
        locateY = data[avgMaxPosition][2]
        
    #    print(data[avgMaxPosition][0])
#        print(total_time)
        resultPoints.append((data[avgMaxPosition][0],data[avgMaxPosition][1],data[avgMaxPosition][2],data[avgMaxPosition][3]))
        data = np.delete(data, avgMaxPosition, 0)
    #  update the parameters
    result_shortest = 0
    value_temp = 0
    locateX_temp = locateX
    locateY_temp = locateY
    result_shortest_points.clear()
    avgMax = 0
    avgMin = 999999
print(sumCount)
print(len(resultPoints))
print(total_time)
# dataframe = pd.DataFrame(resultPoints)
# 将DataFrame存储为csv,index表示是否显示行名，default=True
# dataframe.to_csv(r".\test3.csv", index=False, sep=',')
```
